# Exploratory Data Analysis (EDA)

This notebook explores the processed student dataset to:
- Understand feature distributions
- Identify relationships between study habits and performance
- Validate assumptions before model training

EDA helps ensure that the data is meaningful and suitable for machine learning.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/processed_student_data.csv")
```

```python
df.head()
```

We first examine:
- Shape of the dataset

## Statistical Summary

Descriptive statistics provide insights into:
- Study hours
- Mock test scores
- Revision behavior


```python
df.describe()
```

## Distribution of Study Hours

This visualization helps understand how much time students spend studying daily.

```python
plt.figure()
plt.hist(df["hours_studied_per_day"], bins=20)
plt.xlabel("Study Hours per Day")
plt.ylabel("Number of Students")
plt.title("Distribution of Study Hours")
plt.show()
```

## Mock Test Score Distribution

Mock test scores indicate academic readiness and performance trends.

```python
plt.figure()
plt.hist(df["mock_test_score"], bins=20)
plt.xlabel("Mock Test Score")
plt.ylabel("Number of Students")
plt.title("Distribution of Mock Test Scores")
plt.show()
```

## Study Hours vs Predicted Performance

Analyzing how study time influences predicted performance categories.


```python
plt.figure()
sns.boxplot(x="predicted_performance", y="hours_studied_per_day", data=df)
plt.xlabel("Predicted Performance")
plt.ylabel("Study Hours")
plt.title("Study Hours vs Predicted Performance")
plt.show()
```

## Revision Count vs Predicted Performance

This analysis highlights the importance of revision frequency on performance.


```python
plt.figure()
sns.boxplot(x="predicted_performance", y="revision_count", data=df)
plt.xlabel("Predicted Performance")
plt.ylabel("Revision Count")
plt.title("Revision Count vs Predicted Performance")
plt.show()
```

## Correlation Matrix

Correlation analysis helps identify linear relationships between numerical features.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Matrix")
plt.show()
```

## Key EDA Insights

- Higher study hours generally correlate with better predicted performance
- Mock test score is strongly associated with performance
- Regular revision contributes positively to outcomes
- Data shows no significant missing values or anomalies

These insights support the use of machine learning for personalized study planning.

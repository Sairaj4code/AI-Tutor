# Data Preprocessing

This notebook handles:
- Encoding categorical variables
- Preparing clean feature sets for ML
- Saving processed data for reuse

Keeping preprocessing separate improves modularity and collaboration.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../data/raw/ai_study_companion_dataset.csv")
```

```python
le = LabelEncoder()
```

```python
df["subject"] = le.fit_transform(df["subject"])
df["topic_difficulty"] = le.fit_transform(df["topic_difficulty"])
df["predicted_performance"] = le.fit_transform(df["predicted_performance"])
```

```python
df.to_csv("../data/processed/processed_student_data.csv", index=False)
```

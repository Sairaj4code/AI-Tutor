```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
n_samples = 300
```

```python
subjects = ["Math", "Physics", "Chemistry", "Biology", "Computer Science"]
difficulty_levels = ["Easy", "Medium", "Hard"]
```

```python
data = {
    "student_id": range(1, n_samples + 1),
    "subject": np.random.choice(subjects, n_samples),
    "hours_studied_per_day": np.round(np.random.uniform(0.5, 6.0, n_samples), 1),
    "days_before_exam": np.random.randint(1, 45, n_samples),
    "revision_count": np.random.randint(0, 5, n_samples),
    "topic_difficulty": np.random.choice(difficulty_levels, n_samples),
    "mock_test_score": np.random.randint(30, 100, n_samples),
    "attendance_percentage": np.random.randint(60, 100, n_samples),
}
```

```python
df = pd.DataFrame(data)
```

```python
conditions = [
    (df["mock_test_score"] >= 75) & (df["hours_studied_per_day"] >= 4),
    (df["mock_test_score"] < 50)
]
```

```python
choices = ["Good", "Poor"]
```

```python
df["predicted_performance"] = np.select(
    conditions,
    choices,
    default="Average"
)
```

```python
df.to_csv("ai_study_companion_dataset.csv", index=False)
```
